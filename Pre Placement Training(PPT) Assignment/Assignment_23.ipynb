Question-1:

Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

**Examples :**

Input  : nlnll
Output : 2
Explanation :
        n

       /   \

     l      n

           / \

           l   l

Input  : nlnnlll
Output : 3
          n

       /   \

     l       n

           /    \

           n      l
           /       \
           l        l

```python
def calculateDepth(preorder):
    stack = []
    depth = 0

    for ch in preorder:
        if ch == 'n':
            stack.append(ch)
        elif ch == 'l':
            while stack and stack[-1] == 'l':
                stack.pop()
            if stack:
                stack.pop()
            stack.append('l')
            depth = max(depth, len(stack))

    return depth


# Test the function
preorder = "nlnll"
depth = calculateDepth(preorder)
print("Depth:", depth)
```

Question-2:

Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          /   \

        5     2

             /   \

            3     1

           /  \

          6    7

***Output:** 4 5 3 6*
***Input:***

                    1

                  /   \

                2       3

                 \

                   4

                     \

                        5

                           \

                             6

**Output:** 1 2 4 5 6

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def leftView(root):
    if root is None:
        return

    queue = []
    queue.append(root)

    while queue:
        count = len(queue)

        # Process the leftmost node of the current level
        print(queue[0].data, end=" ")

        # Remove nodes of the current level and enqueue their left and right children
        while count > 0:
            node = queue.pop(0)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

            count -= 1


# Create the binary tree
root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)

# Print the left view of the binary tree
print("Left View: ", end="")
leftView(root)

```

Question-3:

Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def rightView(root):
    if root is None:
        return

    queue = []
    queue.append(root)

    while queue:
        count = len(queue)

        # Traverse all nodes of the current level
        for i in range(count):
            node = queue.pop(0)

            # Print the rightmost node of the current level
            if i == count - 1:
                print(node.data, end=" ")

            # Enqueue the left and right children of the current node
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)


# Create the binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.left.right = Node(8)

# Print the right view of the binary tree
print("Right View: ", end="")
rightView(root)
```

Question-4:

Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output:** 5, 10, 3, 14, 25.

**Input:**

             20

           /     \

        8         22

    /      \      /   \

 5         3    4     25

         /    \

     10       14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

```python
class Node:
    def __init__(self, key):
        self.data = key
        self.left = None
        self.right = None
        self.horizontal_distance = 0


def bottomView(root):
    if root is None:
        return

    # Dictionary to store the bottom view nodes
    bottom_view = {}

    # Queue for level order traversal
    queue = []

    # Initialize the horizontal distance of the root node as 0
    root.horizontal_distance = 0
    queue.append(root)

    while queue:
        node = queue.pop(0)

        # Update the bottom view dictionary with the latest node for each horizontal distance
        bottom_view[node.horizontal_distance] = node.data

        # Enqueue the left and right children of the current node
        if node.left:
            node.left.horizontal_distance = node.horizontal_distance - 1
            queue.append(node.left)
        if node.right:
            node.right.horizontal_distance = node.horizontal_distance + 1
            queue.append(node.right)

    # Print the nodes in the bottom view
    for key in sorted(bottom_view):
        print(bottom_view[key], end=" ")


# Create the binary tree
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

# Print the bottom view of the binary tree
print("Bottom View:")
bottomView(root)
```
